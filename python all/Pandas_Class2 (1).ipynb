## 1 Inspecting Data
Quick ways to understand your dataset:
- `head(n)` → first *n* rows (default 5)
- `tail(n)` → last *n* rows
- `info()` → columns, non-null counts, and dtypes
- `shape` → (rows, columns)
- `dtypes` → data type of each column

```python
import pandas as pd

# Create a small DataFrame
a = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, 30, 35, 40, 22],
    "Score": [88, 92, 79, 85, 95]
}

b = pd.DataFrame(a)
b
```

```python
b.head()
# it will show top 5 rows of the dataframe if ()leave empty otherwise the no.of provided rows
b.head(2)
```

```python
b.tail()
# it will show last 5 rows of the dataframe if ()leave empty otherwise the no.of provided rows
b.tail(2)

```

```python
b.info()
```

```python
b.shape
```

```python
b.dtypes
```

```python
b.describe()
```

## 2 Selecting Columns & Rows
Ways to access parts of a DataFrame:
- Single column → `df['col']`
- Multiple columns → `df[['col1','col2']]`
- Label-based selection → `df.loc[row_label, col_label]`
- Integer-index selection → `df.iloc[row_idx, col_idx]`
Tip: `:` means "all rows/columns" in that dimension.

```python
b
```

```python
b['Name']
```

```python
b[['Name','Score']]
```

```python
b.loc[0]
#loc[lable]
# work same a indexing and select a row index and sho it in column form
```

```python
b.loc[[1,2,3],["Name","Age"]]
```

## 3 Index & Columns Overview
- `df.index` → row labels
- `df.columns` → column names
- `df.values` → underlying NumPy array (read-only; prefer using DataFrame ops)

```python
b.index
# it give the range of index
```

```python
b.columns
# it will provide all the columns present in dataframe
```

```python
b.values
# it will create a 2D array
```

## 4 Renaming Columns & Setting an Index
- Rename columns → `df.rename(columns={'old':'new'}, inplace=False)`
- Set a column as index → `df.set_index('column_name', inplace=False)`
- Reset index → `df.reset_index(inplace=False)`
 If you don't use `inplace=True`, remember to assign the result back to a variable.

```python
b
```

```python
b.rename(columns={"Score":"Marks"},inplace=True)
# {oldname:newname}(you can pass multiple columns), inplace permanently save the change in the main dataframes
b=b.rename(columns={"Score":"Marks"})
# second method is name as first different method
b
```

```python
b.set_index("Name")
# it will change the label
# use inplace=True to save it permanently
```

```python
b.iloc[0]
# if you use set_index and change indexes but you want to access df with normal labels (0,1,2)
```

```python
b.reset_index()
#b.reset_index(inplace=True)
# you canonly run it 2 times only it will make labe(0,1,2,3) as a new column
```

## 5 Basic Attributes Recap
- `df.columns` → names of all columns
- `df.index` → row labels
- `df.values` → NumPy array (data only)
These are handy for quick checks and loops (though vectorized ops are preferred).
